## Funo step

- Funo que retorna o plot dos pontos que esto sendo analisados e da curva do convex hull encontrada at ento. 

```python
def step(pts, c_hull):
    scatter = hv.Scatter(pts)
    curve = hv.Curve(c_hull)
    return scatter * curve
```

## Funo rand_pts

- Funo que retorna um vetor de pontos aleatrios de acordo com  uma distribuio normal.

```python
def rand_pts(mu=0, sigma=0.1, qty=100):
    pts = []
    for i in range(qty):
        s = np.random.normal(mu, sigma, 2)
        pts.append(s)
    return pts
```

## Funo relat_orient

- Funco que retorna a orientao relativa entre dois pontos. 
- Perceba que a resposta  encontrada atravs do clculo do produto vetorial entre dois vetores que comeam no ponto (0,0) e vo, cada um, at os pontos 1 e 2 (pt1 e pt2). 
- Para que seja feita a comparao em relao a um ponto diferente de (0,0),  necessrio passar para os parmetros da funo os pontos 1 e 2 subtrados do ponto de origem p0:  orientao relativa de pt1 e pt2 partindo de p0 = __relat_orient (pt1-p0, pt2-p0)__

```python
def relat_orient(pt1, pt2):    
    c = pt1[0]*pt2[1] - pt2[0]*pt1[1]
    if c > 0:
        return 'direita'
    elif c < 0:
        return 'esquerda'
    else:
        return 'colinear'
```

## Funo find_p0

- Funo que retorna, de uma lista de pontos, o ponto com menor valor em y, e, em caso de empate, com o menor valor em x.

```python
def find_p0(pts):
    p0 = pts[0]
    for i in range(len(pts)):
        if pts[i][1] < p0[1]:
            p0 = pts[i]
        elif pts[i][1] == p0[1]:
            if pts[i][0] < p0[0]:
                p0 = pts[i]
    return p0
```

## Funo sort_by_angle

- Funo que ordena pontos segundo o ngulo polar com relao a p0.

```python

###########        ARRUMAR ISSO:        ###########

#Durante a ordenao, o algoritmo
# mantm somente o vrtice mais
# distante de p0 dentre aqueles que
# possuem a mesma inclinao (ngulo)
# de p0
def sort_by_angle(pts, p0):
    
    ordered_pts = [p0]
    
    for i in range(len(pts)):
        if (pts[i] != p0).any():
            id_min = i
            for j in range(i+1, len(pts)):
                if (pts[j] != p0).any():
                    orient = relat_orient(pts[id_min]-p0, pts[j]-p0)
                    if orient == 'esquerda':
                        id_min = j

            ordered_pts.append(pts[id_min])
            pts[i], pts[id_min] = pts[id_min], pts[i]
            
    return ordered_pts
```

## Funo var_Graham

- Funo que resolve o algoritmo da Varredura de Graham.
- Inicia-se encontrando o ponto de partida p0 atravs da funo find_p0.
- Ordena todos os pontos atravs da funo sort_by_angle.

```python
def var_Graham(pts):
    
    p0 = find_p0(pts)
    
    ordered = sort_by_angle(pts, p0)
        
    conv_h = ordered[:3] 
    
    for i in range(3, len(ordered)):
        while relat_orient(conv_h[-1] - conv_h[-2], ordered[i] - conv_h[-2]) == 'esquerda':
            conv_h = conv_h[:-1]
        conv_h.append(ordered[i])
        
    plt_points(pts, p0)
    plt_points(conv_h, conv_h[0])
    
    
var_Graham(pts)
```

```python
def algorit_inc(pts):
    pts = sorted(pts, key=lambda p : p[0])
    conv_h = pts[:2]
    
    #points neighbours: chave = ponto, [0] = proximo, [1] = anterior
    pts_nbr = {
        0: [1,1],
        1: [0,0]
    }
    
    for i in range(2, len(pts)):
        conv_h.append(pts[i])
        if pts[i][1] > pts[i-1][1]:
            pts_nbr[i] = [i-1,pts_nbr[i-1][1]]
        else:
            pts_nbr[i] = [pts_nbr[i-1][0],i-1]
            
        pts_nbr[pts_nbr[i][0]][1] = i
        pts_nbr[pts_nbr[i][1]][0] = i
        
        next_pt = pts_nbr[i][0]
        next2 = pts_nbr[next_pt][0]
        while relat_orient(pts[next_pt]-pts[i], pts[next2]-pts[i]) == 'direita':
            #vai remover o next_pt do convex hull e atualizar os neighbours
            pts_nbr[i][0] = next2
            pts_nbr[next2][1] = i
            pts_nbr.pop(next_pt)
            to_del = pts[next_pt]
            conv_h = [p for p in conv_h if (p!=to_del).all()]
            
            next_pt = next2
            next2 = pts_nbr[next_pt][0]
    
        prev_pt = pts_nbr[i][1]
        prev2 = pts_nbr[prev_pt][1]
        while relat_orient(pts[prev_pt]-pts[i], pts[prev2]-pts[i]) == 'esquerda':
            #vai remover o prev_pt do convex hull e atualizar os neighbours
            pts_nbr[i][1] = prev2
            pts_nbr[prev2][0] = i
            pts_nbr.pop(prev_pt)
            to_del = pts[prev_pt]
            conv_h = [p for p in conv_h if (p!=to_del).all()]
            
            prev_pt = prev2
            prev2 = pts_nbr[prev_pt][1]
            
    return conv_h    
    

plt_points(pts, pts[0])
conv_h = algorit_inc(pts)
plt_points(conv_h, conv_h[len(conv_h)-1])
```
