```python
import numpy as np
import random
import holoviews as hv
from holoviews import opts
import matplotlib.pyplot as plt

hv.extension('matplotlib')
```

## Funes de plotagem

- A funo step retorna o plot dos pontos que esto sendo analisados e da curva do convex hull encontrada at ento.
- A funo plots insere na lista de step_plots o plot retornado anteriormente e incrementa a contagem de steps.
- Funao stepScatter para plotar s os pontos do passo a passo.
- Funo plt_points plota os pontos com um ponto p0 colocado de outra cor.

```python
def step(c_hull, pts):
    scatter = hv.Scatter(pts)
    curve = hv.Curve(c_hull)
    return scatter * curve
```

```python
def plots(step_plots, count_step, conv_h, pts):
    step_plots.append(step(conv_h, pts)) 
    count_step = count_step + 1
    return step_plots, count_step
```

```python
def stepScatter(c_hull, pts):
    return hv.Scatter(pts).opts(padding=0.1) * hv.Scatter(c_hull).opts(padding=0.1, s=30, color='red')
```

```python
def plotsScatter(step_plots, count_step, conv_h, pts):
    step_plots.append(stepScatter(conv_h, pts)) 
    count_step = count_step + 1
    return step_plots, count_step
```

```python
def plt_points(points, p0):
    plt.scatter([p[0] for p in points], [p[1] for p in points])
    plt.scatter(p0[0], p0[1])
    plt.show()
    
# points = [np.array([4,3]),np.array([1,1]),np.array([1,2]),np.array([2,1]),
#           np.array([0,4]),np.array([3,1.5]),np.array([2.5,3]),np.array([3,2.5])]

#points = [np.array([4,3]),np.array([1,1]),np.array([0,4]),np.array([1.5,3.5]),np.array([3,1.5]), np.array([2.5,2.5])]
```

## Funo rand_pts

- Funo que retorna um vetor de pontos aleatrios de acordo com  uma distribuio normal.

```python
def rand_pts(mu=0, sigma=0.1, qty=100):
    pts = []
    for i in range(qty):
        s = np.random.normal(mu, sigma, 2)
        pts.append(s)
    return pts

pts = rand_pts()
```

## Funo relat_orient

- Funco que retorna a orientao relativa entre dois pontos. 
- Perceba que a resposta  encontrada atravs do clculo do produto vetorial entre dois vetores que comeam no ponto (0,0) e vo, cada um, at os pontos 1 e 2 (pt1 e pt2). 
- Para que seja feita a comparao em relao a um ponto diferente de (0,0),  necessrio passar para os parmetros da funo os pontos 1 e 2 subtrados do ponto de origem p0:  orientao relativa de pt1 e pt2 partindo de p0 = __relat_orient (pt1-p0, pt2-p0)__

```python
def relat_orient(pt1, pt2):    
    c = pt1[0]*pt2[1] - pt2[0]*pt1[1]
    if c > 0:
        return 'direita'
    else:
        return 'esquerda'
```

## Funo find_p0

- Funo que retorna, de uma lista de pontos, o ponto com menor valor em y, e, em caso de empate, com o menor valor em x.

```python
def find_p0(pts):
    p0 = pts[0]
    for i in range(len(pts)):
        if pts[i][1] < p0[1]:
            p0 = pts[i]
        elif pts[i][1] == p0[1]:
            if pts[i][0] < p0[0]:
                p0 = pts[i]
    return p0
```

## Funo sort_by_angle

- Funo que ordena pontos segundo o ngulo polar com relao a p0.

```python
def sort_by_angle(pts, p0):
    
    ordered_pts = [p0]
    
    for i in range(len(pts)):
        if (pts[i] != p0).any():
            id_min = i
            for j in range(i+1, len(pts)):
                if (pts[j] != p0).any():
                    orient = relat_orient(pts[id_min]-p0, pts[j]-p0)
                    if orient == 'esquerda':
                        id_min = j

            ordered_pts.append(pts[id_min])
            pts[i], pts[id_min] = pts[id_min], pts[i]
            
    return ordered_pts
```

## Funo var_Graham

- Funo que resolve o algoritmo da Varredura de Graham.
- Inicia-se encontrando o ponto de partida p0 atravs da funo find_p0.
- Ordena todos os pontos atravs da funo sort_by_angle.
- Os trs primeiros pontos so inseridos no convex hull (conv_h) para iniciar as comparaes
- Ser removido o ltimo ponto do conv_h se ele estiver  esquerda do prximo ponto a ser inserido em relao ao penltimo ponto de conv_h. Isso se repetir at que seja diferente ( direita).

```python
def var_Graham(pts, step_plots=[]):
    
    step_plots.append(step([], pts)) 
    p0 = find_p0(pts)
    ordered = sort_by_angle(pts, p0)    
    conv_h = ordered[:3]
    
    step_plots.append(step(conv_h, pts)) 
    
    for i in range(3, len(ordered)):
        while relat_orient(conv_h[-1] - conv_h[-2], ordered[i] - conv_h[-2]) == 'esquerda':
            conv_h = conv_h[:-1]
        conv_h.append(ordered[i])
        step_plots.append(step(conv_h, pts)) 
    
    conv_h.append(p0)
    step_plots.append(step(conv_h, pts)) 
    
    return step_plots
```

```python
step_plots = var_Graham(pts)

curve_dict = {step:step_plots[step] for step in range(len(step_plots))}
hmap = hv.HoloMap(curve_dict, kdims = 'Step')
hmap
```

## Funo Gift Wrapping

- Funo que resolve o algoritmo do Gift Wrapping
- Inicia-se encontrando o ponto p0 atravs da funo find p_0 e adicionando-o ao conv_h.
- Vai procurar, para cada iterao, o ponto que est mais  esquerda em relao ao ltimo ponto do conv_h e inseri-lo no conv_h.
- O algoritmo acaba quando o ponto mais  esquerda se equivalar a p0.

```python
def gift_wrap(points, step_plots, count_step):
    
    p0 = find_p0(points)
    step_plots, count_step = plots(step_plots, count_step, [], pts)
    
    conv_h = [p0]
    step_plots, count_step = plots(step_plots, count_step, conv_h, pts)
    
    for i in range(len(pts)):
        esq = i
        for j in range(i+1, len(pts)):
            orient = relat_orient(pts[esq]-conv_h[-1], pts[j]-conv_h[-1])
            if orient == 'esquerda':
                esq = j
        if (pts[esq] == p0).all():
            conv_h.append(p0)
            return plots(step_plots, count_step, conv_h, pts)

        conv_h.append(pts[esq])
        step_plots, count_step = plots(step_plots, count_step, conv_h, pts)
        pts[i], pts[esq] = pts[esq], pts[i]
    
```

```python
step_plots, count_step = gift_wrap(pts, step_plots=[], count_step=0)
curve_dict = {step:step_plots[step] for step in range(count_step)}
hmap = hv.HoloMap(curve_dict, kdims = 'Step')
hmap
```

## Funo Algoritmo Incremental

- Funo que resolve o Algoritmo Incremental
- Inicia-se ordenando os pontos em relao  x.
- Adiciona-se os dois primeiros pontos no conv_h e define um dicionrio (pts_nbr) que ser utilizado para se saber os vizinhos dos pontos, sendo eles indicados pelos ndices em que aparecem no vetor de pontos ordenados e, a primeira posio indica os vizinhos no sentido horrio e a segunda no sentido anti-horrio.
-  feita uma iterao pelo vetor de pontos, e, o ponto i sempre far parte do convex hull dessa iterao (j que vai ser o de maior x) mas, para isso, deve-se atualizar quais seus pontos vizinhos (e atualizar, tambm, os seus vizinhos). Sabe-se que o ltimo x inserido sempre ser seu vizinho, e aonde o ponto lido deve ser colocado baseando-se na comparao entre o y desses dois pontos (como prximo no sentido horrio de x se o y do inserido for menor e no anti-horrio se for maior).
- Depois removeremos os pontos vizinhos de p se eles estiverem entre a tangente superior e inferior desse ponto. Isso  feito removendo esse pontos no meio do caminho at que ocorra uma mudana de direo pra direita se estivermos olhando os pontos no sentido horrio e para a esquerda se estivermos olhando no sentido anti-horrio.

```python
def algorit_inc(pts, step_plots, count_step):
    pts = sorted(pts, key=lambda p : p[0])
    
    step_plots, count_step = plotsScatter(step_plots, count_step, [], pts)
    
    conv_h = pts[:2]
    step_plots, count_step = plotsScatter(step_plots, count_step, conv_h, pts)
    
    #points neighbours: chave = ponto, [0] = proximo, [1] = anterior
    pts_nbr = {
        0: [1,1],
        1: [0,0]
    }
    
    for i in range(2, len(pts)):
        
        conv_h.append(pts[i])
        
        if pts[i][1] > pts[i-1][1]:
            pts_nbr[i] = [i-1,pts_nbr[i-1][1]]
        else:
            pts_nbr[i] = [pts_nbr[i-1][0],i-1]
            
        pts_nbr[pts_nbr[i][0]][1] = i
        pts_nbr[pts_nbr[i][1]][0] = i
        
        step_plots, count_step = plotsScatter(step_plots, count_step, conv_h, pts)
        
        next_pt = pts_nbr[i][0]
        next2 = pts_nbr[next_pt][0]
        while relat_orient(pts[next_pt]-pts[i], pts[next2]-pts[i]) == 'direita':
            #vai remover o next_pt do convex hull e atualizar os neighbours
            pts_nbr[i][0] = next2
            pts_nbr[next2][1] = i
            pts_nbr.pop(next_pt)
            to_del = pts[next_pt]
            conv_h = [p for p in conv_h if (p!=to_del).all()]
            
            step_plots, count_step = plotsScatter(step_plots, count_step, conv_h, pts)
            
            next_pt = next2
            next2 = pts_nbr[next_pt][0]
        
        prev_pt = pts_nbr[i][1]
        prev2 = pts_nbr[prev_pt][1]
        while relat_orient(pts[prev_pt]-pts[i], pts[prev2]-pts[i]) == 'esquerda':
            #vai remover o prev_pt do convex hull e atualizar os neighbours
            pts_nbr[i][1] = prev2
            pts_nbr[prev2][0] = i
            pts_nbr.pop(prev_pt)
            to_del = pts[prev_pt]
            conv_h = [p for p in conv_h if (p!=to_del).all()]
            
            step_plots, count_step = plotsScatter(step_plots, count_step, conv_h, pts)
            
            prev_pt = prev2
            prev2 = pts_nbr[prev_pt][1] 
            

    conv_h.append(pts[pts_nbr[len(pts)-1][0]])
    step_plots, count_step = plotsScatter(step_plots, count_step, conv_h, pts)

    return step_plots, count_step   
    
step_plots, count_step = algorit_inc(pts, step_plots=[], count_step=0)
```

```python
curve_dict = {step:step_plots[step] for step in range(count_step)}
hmap = hv.HoloMap(curve_dict, kdims = 'Step')
hmap
```
